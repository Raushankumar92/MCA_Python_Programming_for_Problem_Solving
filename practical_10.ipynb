```python
import math

# Square root
num = float(input("Enter a number to find square root: "))
print("Square root of", num, "is:", math.sqrt(num))

# Factorial
fact_num = int(input("Enter a number to find factorial: "))
print("Factorial of", fact_num, "is:", math.factorial(fact_num))

# Power
base = float(input("Enter base number: "))
exponent = float(input("Enter exponent: "))
print("Result of", base, "^", exponent, "is:", math.pow(base, exponent))
```
