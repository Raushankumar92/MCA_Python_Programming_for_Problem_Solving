```python
# Program to search an item in a user-provided list

# Taking list input from user
items = input("Enter list elements separated by spaces: ").split()

# Item to search
search_item = input("Enter the item to search: ")

# Searching the item
if search_item in items:
    position = items.index(search_item) + 1   # +1 for human-friendly position
    print(f"Item '{search_item}' found at position {position}.")
else:
    print("Item not found.")
```
