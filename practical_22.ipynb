```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("employees_dataset.csv")

# Set seaborn style
sns.set(style="whitegrid")

# 1. Plot a histogram for Salary
plt.figure()
plt.hist(df["Salary"], bins=10)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

# 2. Draw a box plot for Age
plt.figure()
sns.boxplot(y=df["Age"])
plt.title("Age Distribution")
plt.ylabel("Age")
plt.show()

# 3. Create a bar chart showing department-wise employee count
plt.figure()
df["Department"].value_counts().plot(kind="bar")
plt.title("Department-wise Employee Count")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()

# 4. Plot a scatter plot between Experience and Salary
plt.figure()
plt.scatter(df["Experience_Years"], df["Salary"])
plt.title("Experience vs Salary")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.show()

# 5. Display a correlation heatmap
plt.figure()
correlation = df[["Experience_Years", "Salary", "Age"]].corr()
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
