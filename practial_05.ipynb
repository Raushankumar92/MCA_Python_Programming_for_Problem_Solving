```python
#Reverse only vowels in given string.
string = input("Enter a string: ")

vowels = "aeiouAEIOU"
chars = list(string)

# Extract all vowels
vowel_list = [ch for ch in chars if ch in vowels]

# Reverse the vowels
vowel_list = vowel_list[::-1]

# Replace vowels in original positions
j = 0
for i in range(len(chars)):
    if chars[i] in vowels:
        chars[i] = vowel_list[j]
        j += 1

# Join to form final string
result = "".join(chars)
print("Output:", result)
```
