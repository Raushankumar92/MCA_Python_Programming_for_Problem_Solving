```python
# Import pandas
import pandas as pd

# 1. Load the dataset into a pandas DataFrame
df = pd.read_csv("sales_dataset.csv")

print("First five records of the dataset:")
print(df.head())

# 2. Create a new column Total_Revenue
df["Total_Revenue"] = df["Quantity"] * df["Price"]

# 3. Find total revenue generated for each product
print("\nTotal Revenue for each Product:")
print(df.groupby("Product")["Total_Revenue"].sum())

# 4. Find total units sold for each category
print("\nTotal Quantity Sold for each Category:")
print(df.groupby("Category")["Quantity"].sum())

# 5. Calculate the average price per unit for each category
print("\nAverage Price per Unit for each Category:")
print(df.groupby("Category")["Price"].mean())

# 6. Find the product with maximum units sold
max_units_product = df.loc[df["Quantity"].idxmax(), "Product"]
print("\nProduct with Maximum Units Sold:")
print(max_units_product)

# 7. Group the data by Month and calculate total revenue for each month
print("\nTotal Revenue for each Month:")
print(df.groupby("Month")["Total_Revenue"].sum())

# 8. Filter and display products where Quantity is greater than 40
print("\nProducts with Quantity greater than 40:")
print(df[df["Quantity"] > 40])

# 9. Sort the dataset by Total_Revenue in descending order
print("\nDataset sorted by Total Revenue (Descending):")
print(df.sort_values(by="Total_Revenue", ascending=False))
```
