```python
# Program to remove duplicates and display the sorted list

# Input: list with possible duplicates
items = [5, 2, 8, 2, 3, 5, 9, 1, 3]

# Remove duplicates using set
unique_items = list(set(items))

# Sort the list
unique_items.sort()

# Display result
print("Sorted list without duplicates:", unique_items)
```
