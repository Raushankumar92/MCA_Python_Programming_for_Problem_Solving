```python
import random

# List of students
students = ["Aman", "Riya", "Rahul", "Neha", "Suresh", "Pooja"]

# Select CR randomly
cr = random.choice(students)
print("Selected CR:", cr)

# Shuffle the list
random.shuffle(students)

# Print shuffled list
print("Shuffled student list:")
print(students)
```
