```python
rows = 4
cols = 3

for i in range(rows):
    for j in range(cols):
        if i == 0 or i == rows-1 or j == 0 or j == cols-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
rows = 4
for i in range(rows):
    print(" "*(rows-i-1) + "*"*(2*i+1))
```
